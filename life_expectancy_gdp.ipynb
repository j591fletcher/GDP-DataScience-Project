```python
#importing libraries
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter

```

```python
#load data
gdp_data = pd.read_csv('all_data.csv')


#separating data by country
#Chile
chile_year = gdp_data[gdp_data.Country == 'Chile'].Year
chile_life_expectancy = gdp_data[gdp_data.Country == 'Chile']["Life expectancy at birth (years)"]
chile_gdp = gdp_data[gdp_data.Country == 'Chile'].GDP

#China
china_year = gdp_data[gdp_data.Country == 'China'].Year
china_life_expectancy = gdp_data[gdp_data.Country == 'China']["Life expectancy at birth (years)"]
china_gdp = gdp_data[gdp_data.Country == 'China'].GDP

#Germany
germany_year = gdp_data[gdp_data.Country == 'Germany'].Year
germany_life_expectancy = gdp_data[gdp_data.Country == 'Germany']["Life expectancy at birth (years)"]
germany_gdp = gdp_data[gdp_data.Country == 'Germany'].GDP

#Mexico
mexico_year = gdp_data[gdp_data.Country == 'Mexico'].Year
mexico_life_expectancy = gdp_data[gdp_data.Country == 'Mexico']["Life expectancy at birth (years)"]
mexico_gdp = gdp_data[gdp_data.Country == 'Mexico'].GDP

#US
us_year = gdp_data[gdp_data.Country == 'United States of America'].Year
us_life_expectancy = gdp_data[gdp_data.Country == 'United States of America']["Life expectancy at birth (years)"]
us_gdp = gdp_data[gdp_data.Country == 'United States of America'].GDP

#Zimbabwe
zimbabwe_year = gdp_data[gdp_data.Country == 'Zimbabwe'].Year
zimbabwe_life_expectancy = gdp_data[gdp_data.Country == 'Zimbabwe']["Life expectancy at birth (years)"]
zimbabwe_gdp = gdp_data[gdp_data.Country == 'Zimbabwe'].GDP

```

```python
#function to convert GDP in Trillions to a readable format
def trillions_formatter(x, pos):
    return f'{x / 100000000000}'

#creating figure
fig,ax = plt.subplots(figsize=(11,11))

#creating subplots
ax1 = plt.subplot2grid(shape=(3,2), loc=(0,0))
ax2 = plt.subplot2grid(shape=(3,2), loc=(0,1))
ax3 = plt.subplot2grid(shape=(3,2), loc=(1,0))
ax4 = plt.subplot2grid(shape=(3,2), loc=(1,1))
ax5 = plt.subplot2grid(shape=(3,2), loc=(2,0))
ax6 = plt.subplot2grid(shape=(3,2), loc=(2,1))

#First subplot: Chile's GDP 2000-2015

ax1.plot(chile_year, chile_gdp, marker='o')
ax1.set_xticks(chile_year)
ax1.set_xticklabels(chile_year, rotation=45)
ax1.yaxis.set_major_formatter(FuncFormatter(trillions_formatter))
ax1.set_ylabel('GDP (in trillions)')
ax1.grid(axis = 'y')
ax1.set_title('Chile GDP vs. Year')


#Second subplot: China's GDP 2000-2015
ax2.plot(china_year, china_gdp, marker='o')
ax2.set_xticks(china_year)
ax2.set_xticklabels(china_year, rotation=45)
ax2.yaxis.set_major_formatter(FuncFormatter(trillions_formatter))
ax2.set_ylabel('GDP (in trillions)')
ax2.grid(axis = 'y')
ax2.set_title('China GDP vs. Year')


#Third subplot: Germany's GDP 2000-2015
ax3.plot(germany_year, germany_gdp, marker='o')
ax3.set_xticks(germany_year)
ax3.set_xticklabels(germany_year, rotation=45)
ax3.yaxis.set_major_formatter(FuncFormatter(trillions_formatter))
ax3.set_ylabel('GDP (in trillions)')
ax3.grid(axis = 'y')
ax3.set_title('Germany GDP vs. Year')


#Fourth subplot: Mexico's GDP 2000-2015
ax4.plot(mexico_year, mexico_gdp, marker='o')
ax4.set_xticks(mexico_year)
ax4.set_xticklabels(mexico_year, rotation=45)
ax4.yaxis.set_major_formatter(FuncFormatter(trillions_formatter))
ax4.set_ylabel('GDP (in trillions)')
ax4.grid(axis = 'y')
ax4.set_title('Mexico GDP vs. Year')


#Fifth subplot: USA's GDP 2000-2015
ax5.plot(us_year, us_gdp, marker='o')
ax5.set_xticks(us_year)
ax5.set_xticklabels(us_year, rotation=45)
ax5.yaxis.set_major_formatter(FuncFormatter(trillions_formatter))
ax5.set_ylabel('GDP (in trillions)')
ax5.grid(axis = 'y')
ax5.set_title('USA GDP vs. Year')


#Sixth subplot: Zimbabwe's GDP 2000-2015
ax6.plot(zimbabwe_year, zimbabwe_gdp, marker='o')
ax6.set_xticks(zimbabwe_year)
ax6.set_xticklabels(zimbabwe_year, rotation=45)
ax6.yaxis.set_major_formatter(FuncFormatter(trillions_formatter))
ax6.set_ylabel('GDP (in trillions)')
ax6.grid(axis = 'y')
ax6.set_title('Zimbabwe GDP vs. Year')

fig.tight_layout(pad=3.0, w_pad=1.0)
plt.show()
plt.clf()
```

## Analysis
1. The GDP of each country between 2000-2015 has a positive slope indicating the GDP goes up per year.  One noticable trend in GDP in 2009 when it takes a significant dip for each country except for China and Zimbabwe, the later having had a severe dip the year before in 2008.  This may be correlated to the financial crisis in advanced country markets in 2007 which led to a sink of the global economy in 2008-2009.
