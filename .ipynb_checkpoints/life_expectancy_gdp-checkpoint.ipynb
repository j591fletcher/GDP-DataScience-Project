### Questions to this project will answer:
1. How has the GDP of each country changed from 2000 to 2015
2. How is the GDP of a country related to their Life Excetancy


```python
#importing libraries
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
from matplotlib.ticker import FuncFormatter

```

```python
#load data
gdp_data = pd.read_csv('all_data.csv')


#separating data by country
#Chile
chile_year = gdp_data[gdp_data.Country == 'Chile'].Year
chile_life_expectancy = gdp_data[gdp_data.Country == 'Chile']["Life expectancy at birth (years)"]
chile_gdp = gdp_data[gdp_data.Country == 'Chile'].GDP

#China
china_year = gdp_data[gdp_data.Country == 'China'].Year
china_life_expectancy = gdp_data[gdp_data.Country == 'China']["Life expectancy at birth (years)"]
china_gdp = gdp_data[gdp_data.Country == 'China'].GDP

#Germany
germany_year = gdp_data[gdp_data.Country == 'Germany'].Year
germany_life_expectancy = gdp_data[gdp_data.Country == 'Germany']["Life expectancy at birth (years)"]
germany_gdp = gdp_data[gdp_data.Country == 'Germany'].GDP

#Mexico
mexico_year = gdp_data[gdp_data.Country == 'Mexico'].Year
mexico_life_expectancy = gdp_data[gdp_data.Country == 'Mexico']["Life expectancy at birth (years)"]
mexico_gdp = gdp_data[gdp_data.Country == 'Mexico'].GDP

#US
us_year = gdp_data[gdp_data.Country == 'United States of America'].Year
us_life_expectancy = gdp_data[gdp_data.Country == 'United States of America']["Life expectancy at birth (years)"]
us_gdp = gdp_data[gdp_data.Country == 'United States of America'].GDP

#Zimbabwe
zimbabwe_year = gdp_data[gdp_data.Country == 'Zimbabwe'].Year
zimbabwe_life_expectancy = gdp_data[gdp_data.Country == 'Zimbabwe']["Life expectancy at birth (years)"]
zimbabwe_gdp = gdp_data[gdp_data.Country == 'Zimbabwe'].GDP

```

```python
#function to convert GDP in Trillions to a readable format
def trillions_formatter(x, pos):
    return f'{x / 100000000000}'

#creating figure
fig,ax = plt.subplots(figsize=(11,11))

#creating subplots
ax1 = plt.subplot2grid(shape=(3,2), loc=(0,0))
ax2 = plt.subplot2grid(shape=(3,2), loc=(0,1))
ax3 = plt.subplot2grid(shape=(3,2), loc=(1,0))
ax4 = plt.subplot2grid(shape=(3,2), loc=(1,1))
ax5 = plt.subplot2grid(shape=(3,2), loc=(2,0))
ax6 = plt.subplot2grid(shape=(3,2), loc=(2,1))

#First subplot: Chile's GDP 2000-2015

ax1.plot(chile_year, chile_gdp, marker='o')
ax1.set_xticks(chile_year)
ax1.set_xticklabels(chile_year, rotation=45)
ax1.yaxis.set_major_formatter(FuncFormatter(trillions_formatter))
ax1.set_ylabel('GDP (in trillions)')
ax1.grid(axis = 'y')
ax1.set_title('Chile GDP vs. Year')


#Second subplot: China's GDP 2000-2015
ax2.plot(china_year, china_gdp, marker='o')
ax2.set_xticks(china_year)
ax2.set_xticklabels(china_year, rotation=45)
ax2.yaxis.set_major_formatter(FuncFormatter(trillions_formatter))
ax2.set_ylabel('GDP (in trillions)')
ax2.grid(axis = 'y')
ax2.set_title('China GDP vs. Year')


#Third subplot: Germany's GDP 2000-2015
ax3.plot(germany_year, germany_gdp, marker='o')
ax3.set_xticks(germany_year)
ax3.set_xticklabels(germany_year, rotation=45)
ax3.yaxis.set_major_formatter(FuncFormatter(trillions_formatter))
ax3.set_ylabel('GDP (in trillions)')
ax3.grid(axis = 'y')
ax3.set_title('Germany GDP vs. Year')


#Fourth subplot: Mexico's GDP 2000-2015
ax4.plot(mexico_year, mexico_gdp, marker='o')
ax4.set_xticks(mexico_year)
ax4.set_xticklabels(mexico_year, rotation=45)
ax4.yaxis.set_major_formatter(FuncFormatter(trillions_formatter))
ax4.set_ylabel('GDP (in trillions)')
ax4.grid(axis = 'y')
ax4.set_title('Mexico GDP vs. Year')


#Fifth subplot: USA's GDP 2000-2015
ax5.plot(us_year, us_gdp, marker='o')
ax5.set_xticks(us_year)
ax5.set_xticklabels(us_year, rotation=45)
ax5.yaxis.set_major_formatter(FuncFormatter(trillions_formatter))
ax5.set_ylabel('GDP (in trillions)')
ax5.grid(axis = 'y')
ax5.set_title('USA GDP vs. Year')


#Sixth subplot: Zimbabwe's GDP 2000-2015
ax6.plot(zimbabwe_year, zimbabwe_gdp, marker='o')
ax6.set_xticks(zimbabwe_year)
ax6.set_xticklabels(zimbabwe_year, rotation=45)
ax6.yaxis.set_major_formatter(FuncFormatter(trillions_formatter))
ax6.set_ylabel('GDP (in trillions)')
ax6.grid(axis = 'y')
ax6.set_title('Zimbabwe GDP vs. Year')


#Display plot
fig.tight_layout(pad=3.0, w_pad=1.0)
plt.show()
plt.clf()
```

## Analysis
1. The GDP of each country between 2000-2015 has a positive slope indicating the GDP goes up per year.  One noticable trend in GDP in 2009 when it takes a significant dip for each country except for China and Zimbabwe, the later having had a severe dip the year before in 2008.  This may be correlated to the financial crisis in advanced country markets in 2007 which led to a sink of the global economy in 2008-2009.

```python
#creating figure
fig,ax = plt.subplots(figsize=(11,11))

#creating subplots
ax1 = plt.subplot2grid(shape=(3,2), loc=(0,0))
ax2 = plt.subplot2grid(shape=(3,2), loc=(0,1))
ax3 = plt.subplot2grid(shape=(3,2), loc=(1,0))
ax4 = plt.subplot2grid(shape=(3,2), loc=(1,1))
ax5 = plt.subplot2grid(shape=(3,2), loc=(2,0))
ax6 = plt.subplot2grid(shape=(3,2), loc=(2,1))


#First subplot: Chile's life expetency 2000-2015
ax1.plot(chile_year, chile_life_expectancy, marker='o')
ax1.set_xticks(chile_year)
ax1.set_xticklabels(chile_year, rotation=45)
ax1.set_ylabel('Life Expectancy')
ax1.grid(axis = 'y')
ax1.set_title('Chile Life Expectancy 2000-2015')

#Second subplot: China's life expetency 2000-2015
ax2.plot(china_year, china_life_expectancy, marker='o')
ax2.set_xticks(china_year)
ax2.set_xticklabels(china_year, rotation=45)
ax2.set_ylabel('Life Expectancy')
ax2.grid(axis = 'y')
ax2.set_title('China Life Expectancy 2000-2015')


#Third subplot: Germany's life expetency 2000-2015
ax3.plot(germany_year, germany_life_expectancy, marker='o')
ax3.set_xticks(germany_year)
ax3.set_xticklabels(germany_year, rotation=45)
ax3.set_ylabel('Life Expectancy')
ax3.grid(axis = 'y')
ax3.set_title('Germany Life Expectancy 2000-2015')

#Fourth subplot: Mexico's life expetency 2000-2015
ax4.plot(mexico_year, mexico_life_expectancy, marker='o')
ax4.set_xticks(mexico_year)
ax4.set_xticklabels(mexico_year, rotation=45)
ax4.set_ylabel('Life Expectancy')
ax4.grid(axis = 'y')
ax4.set_title('Mexico Life Expectancy 2000-2015')

#Fifth subplot: USA's life expetency 2000-2015
ax5.plot(us_year, us_life_expectancy, marker='o')
ax5.set_xticks(us_year)
ax5.set_xticklabels(us_year, rotation=45)
ax5.set_ylabel('Life Expectancy')
ax5.grid(axis = 'y')
ax5.set_title('USA Life Expectancy 2000-2015')

#Sixth subplot: Zimbabwe's life expetency 2000-2015
ax6.plot(zimbabwe_year, zimbabwe_life_expectancy, marker='o')
ax6.set_xticks(zimbabwe_year)
ax6.set_xticklabels(zimbabwe_year, rotation=45)
ax6.set_ylabel('Life Expectancy')
ax6.grid(axis = 'y')
ax6.set_title('Zimbabwe Life Expectancy 2000-2015')

#Display plot
fig.tight_layout(pad=3.0, w_pad=1.0)
plt.show()
plt.clf()
```

```python
#function to calculate average life expectancy
def calcAvg(country):
    avg = 0
    lifeExpectancy = gdp_data[gdp_data.Country == country]["Life expectancy at birth (years)"]
    denom = 16
    for i in lifeExpectancy:
        avg += i
        
    return avg/denom
    
#creating y values to input into below bar graph
countries = ["Chile", "China", "Germany", "Mexico", "United States of America", "Zimbabwe"]

ybar = []
for i in countries:
    ybar.append(calcAvg(i))
    
#creating bar graph
ax = plt.subplot()
sns.barplot(x=countries,y=ybar)
ax.set_xticklabels(["Chile", "China", "Germany", "Mexico", "USA", "Zimbabwe"])
plt.xticks(rotation=-30)
plt.title("Average Life Expectancy per Country (2000-2015)")
plt.xlabel('Country')
plt.ylabel("Life Expectancy")
ax.bar_label(ax.containers[0])
ax.set_ylim(45,85)

plt.show()
plt.clf()
```

## Analysis
1.  The average life expectancy of the first five countries (Chile, China, Germany, Mexico, USA) is 77.32 which is significantly higher than that of Zimbabwe.  Contributing factors to this exist outside of the dataset analyzed, but the low life expectancy of Zimbabwe may be contributed lack of access to healthcare to treat diseases such as HIV and Malaria.  This is unprovable with the current dataset, so another explanation may be related to the GDP of each country which is explored below.


```python
#creating scatter plot to analyze relationship between Life Expectancy and GDP from 2000-2015

sns.scatterplot(x='GDP', y='Life expectancy at birth (years)', data=gdp_data, hue='Country')
plt.title("GDP vs. Life Expectancy")
```

#### This view is not very helpful as many points are scrunched together and hard to read.  Lets change the view to per country.

```python
#function to convert GDP in Trillions to a readable format
def trillions_formatter(x, pos):
    return f'{x / 100000000000}'

#creating figure
fig,ax = plt.subplots(figsize=(11,11))

#log GDPs
log_chile_gdp = np.log(chile_gdp)
log_china_gdp = np.log(china_gdp)
log_germany_gdp = np.log(germany_gdp)
log_mexico_gdp = np.log(mexico_gdp)
log_us_gdp = np.log(us_gdp)
log_zimbabwe_gdp = np.log(zimbabwe_gdp)

#creating subplots
ax1 = plt.subplot2grid(shape=(3,2), loc=(0,0))
ax2 = plt.subplot2grid(shape=(3,2), loc=(0,1))
ax3 = plt.subplot2grid(shape=(3,2), loc=(1,0))
ax4 = plt.subplot2grid(shape=(3,2), loc=(1,1))
ax5 = plt.subplot2grid(shape=(3,2), loc=(2,0))
ax6 = plt.subplot2grid(shape=(3,2), loc=(2,1))

#First subplot: Chile's GDP vs Life Expectancy
ax1.scatter(chile_gdp, chile_life_expectancy)
ax1.set_xticklabels(log_chile_gdp)
ax1.xaxis.set_major_formatter(FuncFormatter(trillions_formatter))
ax1.set_xlabel('GDP (log value)')
ax1.set_ylabel('Life Expectancy')
ax1.set_title('Chile GDP vs. Life Expectancy')


#Second subplot: China's GDP vs Life Expectancy
ax2.scatter(china_gdp, china_life_expectancy, color='orange')
ax2.set_xticklabels(log_china_gdp)
ax2.xaxis.set_major_formatter(FuncFormatter(trillions_formatter))
ax2.set_xlabel('GDP (log value)')
ax2.set_ylabel('Life Expectancy')
ax2.set_title('China GDP vs. Life Expectancy')

#Third subplot: Germany's GDP vs Life Expectancy
ax3.scatter(china_gdp, germany_life_expectancy, color='green')
ax3.set_xticklabels(log_germany_gdp)
ax3.xaxis.set_major_formatter(FuncFormatter(trillions_formatter))
ax3.set_xlabel('GDP (log value)')
ax3.set_ylabel('Life Expectancy')
ax3.set_title('Germany GDP vs. Life Expectancy')


#Fourth subplot: Mexico's GDP vs Life Expectancy
ax4.scatter(china_gdp, mexico_life_expectancy, color='red')
ax4.set_xticklabels(log_mexico_gdp)
ax4.xaxis.set_major_formatter(FuncFormatter(trillions_formatter))
ax4.set_xlabel('GDP (log value)')
ax4.set_ylabel('Life Expectancy')
ax4.set_title('Mexico GDP vs. Life Expectancy')


#Fifth subplot: USA's GDP vs Life Expectancy
ax5.scatter(china_gdp, us_life_expectancy, color='purple')
ax5.set_xticklabels(log_us_gdp)
ax5.xaxis.set_major_formatter(FuncFormatter(trillions_formatter))
ax5.set_xlabel('GDP (log value)')
ax5.set_ylabel('Life Expectancy')
ax5.set_title('US GDP vs. Life Expectancy')


# #Sixth subplot: Zimbabwe's GDP vs Life Expectancy
ax6.scatter(china_gdp, zimbabwe_life_expectancy, color='brown')
ax6.set_xticklabels(log_zimbabwe_gdp)
ax6.xaxis.set_major_formatter(FuncFormatter(trillions_formatter))
ax6.set_xlabel('GDP (log value)')
ax6.set_ylabel('Life Expectancy')
ax6.set_title('Zimbabwe GDP vs. Life Expectancy')


#Display plot
fig.tight_layout(pad=3.0, w_pad=1.0)
plt.show()
plt.clf()
```

## Analysis
1. These plots indicate a clear correlation between a countries rising GDP and the rising Life Expectancy.  As the countries GDP rises, so does its Life Expectancy.  While not indicated in these plots that are separated by country, we can hypothesize that Zimbabwe's lower Life Expextancy is due to their lower GDP compared to the other 5 countries in the dataset.
